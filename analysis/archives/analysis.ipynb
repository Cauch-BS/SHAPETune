```python
history = open("history.log", "r")
lines = history.readlines()
history.close()
losses = []

for line in lines:
    if "Iteration" in line:
        if "Beginning" not in line:
            split_line = line.split(":")
            losses.append(split_line[-1].strip())
```

```python
min_index = losses.index(min(losses))
print("Minimum loss: ", min(losses))
print("Iteration: ", min_index)
```

```python
import matplotlib.pyplot as plt
losses = [float(i) for i in losses]
plt.plot(losses)
```

```python
grads = []

for line in lines:
    if "Total gradient norm: " in line:
        split_line = line.split(":")
        grads.append(split_line[-1].strip())

grads = [float(i) for i in grads]

plt.plot(grads)
```
