```python
history = open("history_G.log", "r")
lines = history.readlines()
history.close()
losses = []
validation = []

for line in lines:
    if "Iteration" in line:
        if "Beginning" not in line:
            split_line = line.split(":")
            if "loss" in line:
                losses.append(split_line[-1].strip())
            if "valid" in line:
                validation.append(split_line[-1].strip())
```

```python
losses = [float(i) for i in losses]
validation = [float(i) for i in validation]
len(losses), len(validation)
```

```python
min_index = losses.index(min(losses))
min_val = validation[min_index]
print("Minimum loss: ", min(losses))
print("Validation at minimum loss: ", min_val)
print("Iteration: ", min_index)

smal_val = validation.index(min(validation))
smal_loss = losses[smal_val]
print("Loss at minimum validation: ", smal_loss)
print("Minimum validation: ", min(validation))
print("Iteration: ", smal_val)
```

```python
import matplotlib.pyplot as plt
# Create the figure and the first axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the first list on the first axis
color1 = 'tab:blue'
ax1.set_xlabel('Index')
ax1.set_ylabel('Losses', color=color1)
ax1.plot(losses, color=color1, label='List 1')
ax1.tick_params(axis='y', labelcolor=color1)

# Create a twin axis
ax2 = ax1.twinx()

# Plot the second list on the twin axis
color2 = 'tab:orange'
ax2.set_ylabel('Validation Score', color=color2)
ax2.plot(validation, color=color2, label='List 2')
ax2.tick_params(axis='y', labelcolor=color2)

# Add a title
plt.title('Plot of Loss and Validation')

# Add legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Adjust layout and display the plot
fig.tight_layout()
plt.show()
```
