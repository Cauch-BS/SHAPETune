```python
# graph Gamma Distributions given k and theta
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

params = {
    "A0_k": np.exp(0.3146640447),
    "A1_k": np.exp(0.997568854),
    "C0_k": np.exp(0.4536549189),
    "C1_k": np.exp(0.7960273654),
    "G0_k": np.exp(0.3607764847),
    "G1_k": np.exp(0.7175884215),
    "U0_k": np.exp(0.3820697829),
    "U1_k": np.exp(0.6217085081),
    "A0_t": np.exp(-2.118602351),
    "A1_t": np.exp(-0.9497727393),
    "C0_t": np.exp(-2.525121753),
    "C1_t": np.exp(-1.598290642),
    "G0_t": np.exp(-2.065437383),
    "G1_t": np.exp(-0.9349060146),
    "U0_t": np.exp(-2.391501329),
    "U1_t": np.exp(-0.8960615316)
}
```

```python
# Define the distribution parameters to be plotted
names = [
    "A unpaired",
    "A paired",
    "C unpaired",
    "C paired",
    "G unpaired",
    "G paired",
    "U unpaired",
    "U paired",
]
k_values = [
    params["A0_k"],
    params["A1_k"],
    params["C0_k"],
    params["C1_k"],
    params["G0_k"],
    params["G1_k"],
    params["U0_k"],
    params["U1_k"],
]
theta_values = [
    params["A0_t"],
    params["A1_t"],
    params["C0_t"],
    params["C1_t"],
    params["G0_t"],
    params["G1_t"],
    params["U0_t"],
    params["U1_t"],
]
linestyles = ["-", "--"] * 4
colors = [
    "#4c72b0",
    "#a1c9f4",
    "#55a868",
    "#8de5a1",
    "#c44e52",
    "#ff9f9b",
    "#000000",
    "#8c8c8c",
]
x = np.linspace(0, 2.5, 1000)
x = np.linspace(0, 2.5, 1000)
fig, ax = plt.subplots(figsize=(10, 10))

for (k, theta, ls, clr, name) in zip(k_values, theta_values, linestyles, colors, names):
    dist = stats.gamma(k, scale=theta)
    plt.plot(x, dist.pdf(x), ls=ls, c= clr,
             label=f'{name} (k = {k:.2f}, theta = {theta:.2f})')

plt.xlim(0, 2.5)
plt.ylim(0, 6)
plt.legend()
plt.show()
```
